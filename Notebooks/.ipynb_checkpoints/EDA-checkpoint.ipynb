```python
# First job is to import the required libraries and the data that will be all our csv files from the dataset

import pandas as pd

orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
```

```python
# Understanding the data is the first step
orders.info()
orders.head()
```

```python
orders.tail()
```

```python
# Converting all date data from object datatype to date datatype
date_columns = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date',
                'order_estimated_delivery_date']

for columns in date_columns:
    orders[columns] = pd.to_datetime(orders[columns], errors='coerce')
orders.info()
```

```python
# Keep missing delivery dates (important for analysis)
# Fill product categories
products['product_category_name'].fillna('unknown', inplace=True)

# Reviews text can be empty
reviews['review_comment_message'].fillna('', inplace=True)
```

```python
# Removing duplicates
orders = orders.drop_duplicates()
customers = customers.drop_duplicates()
order_items = order_items.drop_duplicates()
```

```python
# Fixing datatypes
order_items['price'] = order_items['price'].astype(float)
order_items['freight_value'] = order_items['freight_value'].astype(float)

payments['payment_value'] = payments['payment_value'].astype(float)
```

```python
# Standerdise column names
def clean_columns(df):
    df.columns = df.columns.str.lower().str.strip()
    return df

orders = clean_columns(orders)
customers = clean_columns(customers)
```

```python
# Feature engineering
orders['delivery_time'] = (
    orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
).dt.days

orders['delay'] = (
    orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']
).dt.days
```

```python
order_total = order_items.groupby('order_id')['price'].sum().reset_index()
order_total.rename(columns={'price': 'total_product_value'}, inplace=True)
```

```python
# Merging tables
df = orders.merge(customers, on='customer_id', how='left') \
           .merge(order_items, on='order_id', how='left') \
           .merge(products, on='product_id', how='left') \
           .merge(payments, on='order_id', how='left') \
           .merge(reviews, on='order_id', how='left')
```

```python
# Handling inconsistencies
payment_check = payments.groupby('order_id')['payment_value'].sum()
```

```python
# Clean files
orders.to_csv("clean_orders.csv", index=False)
customers.to_csv("clean_customers.csv", index=False)
order_items.to_csv("clean_order_items.csv", index=False)
```
